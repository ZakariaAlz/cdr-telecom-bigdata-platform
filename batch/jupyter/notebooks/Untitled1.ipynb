```python
# In your Jupyter notebook:
from pyspark.sql import SparkSession

# Initialize Spark with Delta Lake support
spark = SparkSession.builder \
    .appName("CDR Analysis") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://namenode:9000") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

# Read data from HDFS
cdr_data = spark.read.csv("hdfs://namenode:9000/user/data/raw/converted", header=True, inferSchema=True)

# Save in Delta format
cdr_data.write.format("delta").mode("overwrite").save("hdfs://namenode:9000/user/data/processed/cdr_delta")

# Create database and table
spark.sql("CREATE DATABASE IF NOT EXISTS delta_db")
spark.sql("""
    CREATE TABLE IF NOT EXISTS delta_db.cdr_delta
    USING DELTA
    LOCATION 'hdfs://namenode:9000/user/data/processed/cdr_delta'
""")

# Now you can query this table using SQL
result = spark.sql("SELECT * FROM delta_db.cdr_delta LIMIT 5")
result.show()
```
